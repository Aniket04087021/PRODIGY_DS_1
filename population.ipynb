```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample dataset
np.random.seed(42)  # For reproducibility

# Generate a dataset with 'Age' and 'Gender'
num_samples = 500
ages = np.random.randint(1, 100, num_samples)  # Ages between 1 and 99
genders = np.random.choice(['Male', 'Female', 'Other'], num_samples, p=[0.45, 0.45, 0.1])

# Create a DataFrame
data = pd.DataFrame({'Age': ages, 'Gender': genders})

# Bar Chart: Count by Gender
gender_counts = data['Gender'].value_counts()

# Histogram: Distribution of Ages
plt.figure(figsize=(14, 6))

# Subplot 1: Bar Chart
plt.subplot(1, 2, 1)
gender_counts.plot(kind='bar', color=['skyblue', 'pink', 'purple'], alpha=0.8, edgecolor='black')
plt.title('Gender Distribution', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 2: Histogram
plt.subplot(1, 2, 2)
plt.hist(data['Age'], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Age Distribution', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display plots
plt.tight_layout()
plt.show()
```
